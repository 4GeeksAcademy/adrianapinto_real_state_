#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
# TODO
import pandas as pd
# Este archivo CSV contiene puntos y comas en lugar de comas como separadores
ds = pd.read_csv('assets/real_estate.csv', sep=';')
ds

#["price,""level5"]


sample_poblacion = ds[ds["level5"] == "Arroyomolinos (Madrid)"]
sample_poblacion
print (sample_poblacion)


media_poblacion = sample_poblacion["price"].mean()

print(f"La media de precios en Arroyomolinos (Madrid) es de {media_poblacion} Euros")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')
ds

sample_poblacion = ds[ds["level5"] == "Arroyomolinos (Madrid)"]
sample_poblacion

plt.hist(sample_poblacion["price"], bins = 30, alpha = 0.7)

plt.title("Histograma")
plt.show()

# Trazar el histograma de precios
```

se observa que los precios mas frecuentes estan en un rango entre 230.000 a 350.000 euros aprox.
